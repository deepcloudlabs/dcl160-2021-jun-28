```python
employees = [
    ("jack shephard", "Sales", 100000, 1978,True),
    ("kate austen", "IT", 200000, 1985,True),
    ("ben linus", "Finance", 150000, 1967,True),
    ("james sawyer", "HR", 70000, 1970,True),
    ("kim kwon", "Sales", 120000, 1986,True),
    ("sun kwon", "IT", 200000, 1969,False),
    ("hugo reyes", "IT", 120000, 1992,True)
]
```

```python
# data structure (memory) -> persistence (file system -> files)
# 1) text/binary             i) text io -> text editor -> HRF : json, xml, csv, ... 
#                                                         binary -- base64-encoding --> text 
#                           ii) binary io -> A2A Integration: mp4, mp3, jpeg, png, ...
#
# 2) write, read, append-only, read/write, ...                 
```

```python
try:
  with open("employees.csv", mode="w") as myfile:
    for fullname, department, salary, birth_year, is_full_time in employees:
      # text io -> HRF
      myfile.write(f"{fullname},{salary},{department},{birth_year},{'FULL_TIME' if is_full_time else 'PART_TIME'}\n")
except PermissionError:
  print("Cannot open/write to file. Try to change permission: chmod +w")
except Exception as err:
  print("Error has occured: "+str(err)) 
```

```python
calisanlar=[]
try:
  with open("employees.csv", mode="r") as myfile:
    for line in myfile.readlines():
      fullname, salary, department, birth_year, is_full_time = line.strip().split(",")
      fullname = fullname.title()
      salary = float(salary)
      birth_year = int(birth_year)
      is_full_time = is_full_time == "FULL_TIME"
      calisanlar.append((fullname,salary,birth_year,department,is_full_time)) 
except PermissionError:
  print("Cannot open/write to file. Try to change permission: chmod +w")
except Exception as err:
  print("Error has occured: "+str(err)) 
for calisan in calisanlar:
  print(calisan)
```
