```python
number=50
if number < 0:
  print(f"{number} is negative.")
elif number == 0:
  print(f"{number} is zero.")
elif 0 < number < 50:
  print(f"{number} is between 1 and 49.")
elif number == 50:
  print(f"{number} is fifty.")
else:
  print(f"{number} is larger than 50.")  
```

```python
year=input("Enter a year:")
year = int(year)
if year%4 != 0:
  print(f"{year} is not a leap year")
elif year%100 != 0:
  print(f"{year} is a leap year")
elif year%400 != 0:
  print(f"{year} is not a leap year")
else:
  print(f"{year} is a leap year")
```

```python
sequence=[10,20,30,None,50,None,60,70,None,None,100]
```

```python
values=[]
for seq in sequence:
  if seq is not None:
    values.append(seq)
print(values)    
```

```python
numbers=[10,20,30,None,50,None,60,70,None,None,100]
while None in numbers:
  numbers.remove(None)
print(numbers)  
```

```python
numbers=[10,20,30,None,50,None,60,70,None,None,100]
values=[val for val in numbers if val is not None]
```

```python
for v in values:
  print(v)
```

```python
list(filter(lambda u : u is not None , numbers))   # functional programming
```

```python
list(filter( None , numbers)) # functional programming: filter/map/reduce -> sum
```

```python
sum(map(lambda v : v*v, filter( None , numbers)))
```
