```python
employees = [
    ("jack shephard", "Sales", 100000, 1978,True),
    ("kate austen", "IT", 200000, 1985,False),
    ("ben linus", "Finance", 150000, 1967,True),
    ("james sawyer", "HR", 70000, 1979,True),
    ("kim kwon", "Sales", 120000, 1986,True),
    ("sun kwon", "IT", 170000, 1984,False),
    ("hugo reyes", "IT", 120000, 1992,True)
]
salariesByDepartments = {}
for emp in employees:
  fullname, department,salary,*rest = emp
  if department not in salariesByDepartments:
    salariesByDepartments[department] = { "min": float('inf'), "max": float('-inf'), "total": 0}
  salariesByDepartments[department]["total"] += salary
  if salary < salariesByDepartments[department]["min"]:
    salariesByDepartments[department]["min"] = salary
  if salary > salariesByDepartments[department]["max"]:
    salariesByDepartments[department]["max"] = salary
print(salariesByDepartments)    
```

```python
# 17 -> 52 -> 26 -> 13 -> 40 -> 20 -> 10 -> 5 -> 16 -> 8 -> 4 -> 2 -> (1)
n=19
step=0
sequence=[]
while n > 1:
  sequence.append(n)
  if n%2 == 0:
    n = n//2
  else:
    n = 3 * n + 1
  step = step + 1
print(step)     
print(sequence) 
```

```python
employees = [
    ("jack shephard", "Sales", 100000, 1978,True),
    ("kate austen", "IT", 200000, 1985,False),
    ("ben linus", "Finance", 150000, 1967,True),
    ("james sawyer", "HR", 70000, 1979,True),
    ("kim kwon", "Sales", 120000, 1986,True),
    ("sun kwon", "IT", 170000, 1984,False),
    ("hugo reyes", "IT", 120000, 1992,True)
]
sum(map(lambda emp : 1 if emp[4] else 0, employees))
employeeTypes = { 'FULL-TIME': 0, 'PART-TIME': 0}
for emp in employees:
  job_type = 'FULL-TIME' if emp[4] else 'PART-TIME'
  employeeTypes[job_type] = employeeTypes[job_type] + 1
print(employeeTypes)  
```
