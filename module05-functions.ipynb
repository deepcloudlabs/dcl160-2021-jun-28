```python
employees = [
    ("jack shephard", "Sales", 100000, 1978,True),
    ("kate austen", "IT", 200000, 1985,False),
    ("ben linus", "Finance", 150000, 1967,True),
    ("james sawyer", "HR", 70000, 1979,True),
    ("kim kwon", "Sales", 120000, 1986,True),
    ("sun kwon", "IT", 170000, 1984,False),
    ("hugo reyes", "IT", 120000, 1992,True)
]
```

```python
def get_total_salary(emps, department = 'IT'):
  """
  Takes employees and department name returns total salaries in that department
  """  
  return sum(map(lambda emp : float(emp[2]) , filter(lambda emp : emp[1] == department, emps)))
```

```python
type(get_total_salary)
```

```python
get_total_salary(employees, 'Sales')
```

```python
get_total_salary(employees)
```

```python
def get_departments(emps):
  """
  Takes employees and returns distinct department set
  """
  depts = []
  for emp in emps:
    depts.append(emp[1])
  return set(depts)  
```

```python
for department in get_departments(employees):
  total = get_total_salary(employees, department)
  print(f"{department}:\t{total}")
```

```python
# n is 17 -> [17,52,26,13,40,20,10,5,16,8,4,2,1]
def sun(n):
  seq=[n]
  while n > 1:
    print(f"sun is working: {n}")
    n = n//2 if n%2==0 else 3*n+1
    seq.append(n)
  return seq  
```

```python
sun(17)
```

```python
total=0
for val in sun(17):
  total += val
print(total)
```

```python
def run(n): # generator -> lazy
  while n > 1:
    print(f"run is working: {n}")
    n = n//2 if n%2==0 else 3*n+1
    yield n
```

```python
type(run)
```

```python
seq = run(17)
```

```python
type(seq)
```

```python
total=0
for val in run(17):
  print(f"sub total is {total}")
  total += val
print(total)
```
