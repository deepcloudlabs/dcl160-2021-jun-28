```python
import random
```

```python
# lottery numbers: 6 distinct and sorted numbers between 1-60
lottery_numbers = [] # list is mutable
while len(lottery_numbers) < 6 :
  number = random.randint(1,60)
  if number not in lottery_numbers:
    lottery_numbers.append(number)
lottery_numbers.sort()
print(lottery_numbers)
```

```python
numbers = set([]) # set is mutable
while len(numbers) < 6 :
  numbers.add(random.randint(1,60))
sorted_numbers = sorted(numbers)
print(sorted_numbers)
```

```python
orders = ( ["orcl", 10, 123.47], ["ibm", 120, 73.17],["orcl", 20, 124.38],["ibm", 20, 72.58] )
```

```python
volumes = {} # volumes is a dictionary
for order in orders:
  symbol, price, quantity = order # unpacking list/tuple/set
  volume = price * quantity
  if symbol not in volumes:
    volumes[symbol] = volume
  else:
    volumes[symbol] = volumes[symbol] + volume
```

```python
for symbol in volumes:
  print(f"{symbol}:\t{volumes[symbol]}")
```

```python
for symbol,volume in volumes.items():
  print(f"{symbol}:\t{volume}")
```

```python
total_volume=0
for volume in volumes.values():
  total_volume += volume
print(f"Total stock exchange volume is {total_volume}")  
```
