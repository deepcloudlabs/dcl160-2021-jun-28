```python
numbers = (4,8,15,16,23,42)
```

```python
# procedural programming
total=0
for num in numbers:
  if num%2 != 0:
    squared = num ** 2
    total += squared
print(total)    
```

```python
# functional programming : filter/map/reduce -> parallel programming framework
def is_even_fun(n): # (1)
  print(f"is_even_fun({n}) is working hard!")
  return n%2 == 0


def square_fun(x):
  print(f"square_fun({x}) is working harder!")
  return x**2

# functional programming
#   i) generator function -> lazy evaluation
#  ii) pure/high-order function -> âœ˜ side effect
# iii) immutable data structures

is_even = lambda n : n%2 == 0 # (2)
result = sum(map(square_fun, filter( is_even_fun , numbers))) # generator function -> lazy
# (4,8,15,16,23,42) -> filter -> map -> reduce
#                        4        4       (16)
#                        8        8       (16+64)
#                        15
#                        16      16       (16+64+256)
#                        23
#                        42      42       (16+64+256+1764) -> 2100
# filter( lambda n : n%2 == 0 , numbers ) # (3)
# filter( is_even_fun , numbers)

```

```python
from functools import reduce
```

```python
def sum_fun(acc,n):
   print(f"sum_fun({acc},{n}) is working hard!")
   return acc+n
reduce(sum_fun , map(square_fun, filter( is_even_fun , numbers)), 0)
```

```python
# find the minimum element of the list 'numbers'
reduce(lambda acc,u: acc if acc < u else u , numbers, float('inf') )
reduce(lambda acc,u: acc if acc < u else u , numbers, numbers[0] )
reduce(lambda acc,u: acc if acc < u else u , numbers, numbers[-1] )
```

```python
# find the maximum element of the list 'numbers'
reduce(lambda acc,u: acc if acc > u else u , numbers, float('-inf') )
```

```python
# find the sum of the list 'numbers'
reduce(lambda acc,u: acc + u , numbers, 0 )
```

```python
# dictionary {"odd": 2, "even": 4}
def even_odd_counter(d,odd_even):
  d[odd_even] += 1
  d["total"] += 1
  return d
even_odd_mapper = lambda n : "even" if n%2 == 0 else "odd"
initial_dict = {"odd": 0, "even": 0, "total": 0}
reduce( even_odd_counter , map( even_odd_mapper , numbers), initial_dict )
```

```python
employees = [
    ("jack shephard", "Sales", 100000, 1978,True),
    ("kate austen", "IT", 200000, 1985,True),
    ("ben linus", "Finance", 150000, 1967,True),
    ("james sawyer", "HR", 70000, 1970,True),
    ("kim kwon", "Sales", 120000, 1986,True),
    ("sun kwon", "IT", 200000, 1969,False),
    ("hugo reyes", "IT", 120000, 1992,True)
]
```

```python
reduce(lambda counter,emp : counter+1 , filter( lambda emp : emp[3] < 1971 , employees ) , 0)
```

```python
reduce(lambda counter,n : counter+n , map( lambda emp : 1, filter( lambda emp : emp[3] < 1971 , employees )) , 0)
```

```python
def distinct_reducer( set_acc , value) : 
  set_acc.update([value])
  return set_acc


def dept_dict_reducer(dept_dict,dept):
  dept_dict[dept] = 0
  return dept_dict

emp2dept = lambda emp : emp[1]
departments = reduce( distinct_reducer , map( emp2dept ,employees ) , set([]) )
dept_dict = reduce( dept_dict_reducer , departments , {} ) 
```

```python
dept_dict = reduce( dept_dict_reducer , reduce( distinct_reducer , map( emp2dept ,employees ) , set([]) ) , {} )
```

```python
def counter_reducer(d,dept):
  d[dept] +=1
  return d


reduce( counter_reducer , map(emp2dept,employees) , dept_dict )
```

```python
reduce( counter_reducer , map(emp2dept,employees) , reduce( dept_dict_reducer , reduce( distinct_reducer , map( emp2dept ,employees ) , set([]) ) , {} ) )
```
